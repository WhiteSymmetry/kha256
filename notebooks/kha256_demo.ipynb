```python
import secrets
import struct  # Uzunluk prefix iÃ§in

salt = secrets.token_bytes(32)
original_data = b"Onemli gizli veri"

# Prefix'li hash tag (32 byte hedef)
tag_input = original_data + salt
security_tag_str = hasher.hash(tag_input, salt)  # hex str
security_tag_bytes = bytes.fromhex(security_tag_str[:64])  # Ä°lk 256-bit'i bytes'a (32 byte)

# Sakla: salt(32) + data_len(4) + data + tag(32)
data_len = struct.pack('>I', len(original_data))
stored = salt + data_len + original_data + security_tag_bytes

# Recovery
recovered_salt = stored[:32]
recovered_len = struct.unpack('>I', stored[32:36])[0]
recovered_data_start = 36
recovered_data = stored[recovered_data_start : recovered_data_start + recovered_len]
recovered_tag = stored[-(32):]

# Yeniden hesapla
computed_tag_str = hasher.hash(recovered_data + recovered_salt, recovered_salt)
computed_tag_bytes = bytes.fromhex(computed_tag_str[:64])

print(computed_tag_bytes == recovered_tag)  # True olmalÄ±
print(len(security_tag_str))  # Kontrol: hex uzunluk?
```

```python
print("Orijinal input len:", len(original_data + salt))
print("Recovered input len:", len(recovered_data + recovered_salt))
print("Tag types:", type(security_tag_bytes), type(recovered_tag))
print("Tags equal raw:", security_tag_bytes == recovered_tag)
```

```python
from kha256 import Kha256SecureStorage, FortifiedKhaHash256, FortifiedConfig
import secrets
import struct

class Kha256SecureStorage:
    def __init__(self):
        self.config = FortifiedConfig(iterations=3, shuffle_layers=2, salt_length=32, double_hashing=True)
        self.hasher = FortifiedKhaHash256(self.config)
    
    def protect(self, data: bytes) -> bytes:
        """Veriyi salt + len + data + tag ile korur"""
        salt = secrets.token_bytes(32)
        tag_input = data + salt
        tag_str = self.hasher.hash(tag_input, salt)
        tag_bytes = bytes.fromhex(tag_str[:64])  # 32-byte tag
        
        data_len = struct.pack('>I', len(data))
        return salt + data_len + data + tag_bytes
    
    def verify(self, stored: bytes) -> tuple[bool, bytes]:
        """BÃ¼tÃ¼nlÃ¼ÄŸÃ¼ doÄŸrular, veri dÃ¶ner"""
        salt = stored[:32]
        data_len = struct.unpack('>I', stored[32:36])[0]
        data_start = 36
        data = stored[data_start:data_start + data_len]
        tag = stored[-(32):]
        
        computed_tag_str = self.hasher.hash(data + salt, salt)
        computed_tag = bytes.fromhex(computed_tag_str[:64])
        
        return computed_tag == tag, data

# KullanÄ±m
storage = Kha256SecureStorage()
protected = storage.protect(b"Onemli gizli veri")
is_valid, recovered = storage.verify(protected)
print(f"Valid: {is_valid},\nData: {recovered}")  
# Valid: True, Data: b'Onemli gizli veri'
```

```python
def plot_avalanche_simple(bit_diffs):
    """Basit avalanche grafiÄŸi"""
    import matplotlib.pyplot as plt
    import numpy as np
    
    plt.figure(figsize=(15, 5))
    
    # Histogram
    plt.subplot(1, 2, 1)
    plt.hist(bit_diffs, bins=25, alpha=0.7, color='steelblue', edgecolor='black')
    plt.axvline(128, color='red', linestyle='--', linewidth=2, label='Ä°deal (128)')
    plt.axvline(np.mean(bit_diffs), color='darkgreen', linestyle='-', linewidth=2, 
                label=f'Ortalama ({np.mean(bit_diffs):.2f})')
    plt.xlabel('FarklÄ± Bit SayÄ±sÄ±')
    plt.ylabel('Frekans')
    plt.title('Avalanche DaÄŸÄ±lÄ±mÄ±')
    plt.legend()
    plt.grid(alpha=0.3)
    
    # Box plot
    plt.subplot(1, 2, 2)
    bp = plt.boxplot(bit_diffs, vert=True, patch_artist=True)
    bp['boxes'][0].set_facecolor('lightblue')
    plt.axhline(128, color='red', linestyle='--', label='Ä°deal')
    plt.axhline(np.mean(bit_diffs), color='green', linestyle='-', label='Ortalama')
    plt.ylabel('FarklÄ± Bit SayÄ±sÄ±')
    plt.title('Kutu GrafiÄŸi')
    plt.legend()
    plt.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# KullanÄ±m
plot_avalanche_simple(bit_diffs)
```

```python
import kha256
from kha256 import FortifiedKhaHash256, quick_hash
import secrets

# Global salt
GLOBAL_SALT = secrets.token_bytes(32)

def unicode_hash(metin, log=True):
    data = metin.encode('utf-8')
    
    # HÄ±zlÄ± mod (log'suz) - quick_hash saltsÄ±z
    if not log:
        return quick_hash(data)  # âœ… saltsÄ±z quick_hash
    
    # Tam fortified mod - saltlÄ±
    hasher = FortifiedKhaHash256()
    result = hasher.hash(data, salt=GLOBAL_SALT)
    print("[KHA-256 Ã‡Ä±ktÄ±sÄ±]")
    return result

# Test
print(unicode_hash("Ã§", log=False))      # HÄ±zlÄ±, saltsÄ±z
print(unicode_hash("Merhaba â˜º"))         # Fortified, saltlÄ±
```
